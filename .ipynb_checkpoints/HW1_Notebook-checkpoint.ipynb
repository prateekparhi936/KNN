```python
<br><br>
### 1. Vertebral Column Data Set
<p>
This Biomedical data set was built by Dr. Henrique da Mota during a medical residence
period in Lyon, France. Each patient in the data set is represented in the data set
by six biomechanical attributes derived from the shape and orientation of the pelvis
and lumbar spine (in this order): pelvic incidence, pelvic tilt, lumbar lordosis angle,
sacral slope, pelvic radius and grade of spondylolisthesis. The following convention is
used for the class labels: DH (Disk Hernia), Spondylolisthesis (SL), Normal (NO) and
Abnormal (AB). In this exercise, we only focus on a binary classification task NO=0
and AB=1.
</p>
```

```python
**(a)** Download the Vertebral Column Data Set from:https://archive.ics.uci.
edu/ml/datasets/Vertebral+Column
```

```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import seaborn as sn
import matplotlib.pyplot as plt
sn.set(style='ticks')


def preprocess_data():
    column_names=['pelvic incidence','pelvic tilt','lumbar lordosis angle','sacral slope',
                  'pelvic radius','grade of spondylolisthesis','Class']
    Data=pd.read_csv('https://raw.githubusercontent.com/prateekparhi936/HW1-Dataset/master/vertebral_column_data/column_2C.dat',sep='\s+',names=column_names,header=None)
    for i,val in enumerate(Data['Class']):
        if val=='AB':
            Data.at[i,'Class']=1
        elif val=='NO':
            Data.at[i,'Class']=0
    return Data


Data = preprocess_data()
print("\n\n")
```

**III.** Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set and
the rest of the data as the test set.

```python
def split_train_test(): 
    rowClass0=70
    rowClass1=140
    noCol=7
    trainClass0=np.zeros(shape=(rowClass0,noCol))
    DataClass0=[]
    for i,row in Data.iterrows():
        if(row['Class']==0):
            DataClass0.append(row)
    for i,row in enumerate(DataClass0):
        if i==rowClass0:
            break
        trainClass0[i]=row
    testClass0=np.zeros(shape=(len(DataClass0)-trainClass0.shape[0],noCol))
    x=trainClass0.shape[0]
    for i,row in enumerate(DataClass0):
        if i==len(DataClass0):
            break
        if i>=x:
            testClass0[x-i]=row
    trainClass0=pd.DataFrame(trainClass0)
    trainClass0.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle',
                         'sacral slope', 'pelvic radius', 'grade of spondylolisthesis', 'Class']
    trainClass0.Class=trainClass0.Class.astype(int)
    testClass0=pd.DataFrame(testClass0)
    testClass0.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope',
                        'pelvic radius', 'grade of spondylolisthesis', 'Class']
    testClass0.Class=testClass0.Class.astype(int)
    trainClass1=np.zeros(shape=(rowClass1,noCol))
    DataClass1=[]
    for i,row in Data.iterrows():
        if(row['Class']==1):
            DataClass1.append(row)
    for i,row in enumerate(DataClass1):
        if i==rowClass1:
            break
        trainClass1[i]=row
    testClass1=np.zeros(shape=(len(DataClass1)-trainClass1.shape[0],noCol))
    x=trainClass1.shape[0]
    for i,row in enumerate(DataClass1):
        if i==len(DataClass1):
            break
        if i>=x:
            testClass1[x-i]=row
    testClass1=pd.DataFrame(testClass1)
    testClass1.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope', 
                        'pelvic radius', 'grade of spondylolisthesis', 'Class']
    testClass1.Class=testClass1.Class.astype(int) # converts 1.1 to 1
    trainClass1=pd.DataFrame(trainClass1)
    trainClass1.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope', 
                         'pelvic radius', 'grade of spondylolisthesis', 'Class']
    trainClass1.Class=trainClass1.Class.astype(int)
    DataTrain=pd.concat([trainClass0,trainClass1],ignore_index=True)
    X_train=DataTrain[['pelvic incidence','pelvic tilt','lumbar lordosis angle',
                       'sacral slope','pelvic radius','grade of spondylolisthesis']]
    Y_train=DataTrain['Class']
    Y_train=Y_train.astype('int')
    DataTest=pd.concat([testClass0,testClass1],ignore_index=True)
    X_test=DataTest[['pelvic incidence','pelvic tilt',
                     'lumbar lordosis angle','sacral slope','pelvic radius','grade of spondylolisthesis']]
    Y_test=DataTest['Class']
    Y_test=Y_test.astype('int')   
    return X_train,Y_train,X_test,Y_test,DataTrain,DataTest



X_train,Y_train,X_test,Y_test,DataTrain,DataTest=split_train_test()
print("\n\n")
```

**(c)** Classification using KNN on Vertebral Column Data Set<br><br>
**I.** Write code for k-nearest neighbors with Euclidean metric (or use a software package)

```python
import math
from sklearn.neighbors import KNeighborsClassifier
from numpy import empty
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from texttable import Texttable

train_error1=np.empty([105])
K_neighbors=np.arange(210,0,-2)


def KNN_eucledian():
    test_error1=np.empty([105])
    maxScore=0
    most_suitable_k=0
    score=0
    lowest_train_error=math.inf
    for i,k in enumerate(K_neighbors):
        knn_model=KNeighborsClassifier(n_neighbors=k,weights='uniform',algorithm='auto',p=2) #Eucledian Metrics
        knn_model.fit(X_train,Y_train)
        Y_predict=knn_model.predict(X_test)
        score=knn_model.score(X_test,Y_test) #Accuracy
        train_error1[i]=(1-knn_model.score(X_train,Y_train))*100
        test_error1[i]=(1-knn_model.score(X_test,Y_test))*100
        if(score > maxScore):
            maxScore=score
            most_suitable_k=k 
    table1=Texttable()
    table1.add_rows([['Most Suitable K','Maximum Score achieved'],[most_suitable_k,maxScore]])
    print(table1.draw(),"\n")
    return most_suitable_k,maxScore,train_error1,test_error1


most_suitable_k,maxScore,train_error1,test_error1=KNN_eucledian()
print("\n\n")
```

**II.** Test all the data in the test database with k nearest neighbors. Take decisions by
    majority polling. Plot train and test errors in terms of k for k = (208 , 205,.,7, 4, 1) (in
    reverse order). You are welcome to use smaller increments of k. Which k is the most
    suitable k among those values? Calculate the confusion matrix, true positive rate, true
    negative rate, precision and F-score when k = k


```python
def plot_train_test_error(train_error1,test_error1):
    print("\n Train Error vs Test Error Plot")
    plt.title("Plotting training vs testing error with varying Neighbors(k)")
    plt.plot(K_neighbors,train_error1,label='train error',color='red')
    plt.plot(K_neighbors,test_error1,label='test error',color='blue')
    plt.legend()
    plt.xlabel("NEIGHBORS(K)")
    plt.ylabel("ERROR RATES")
    plt.show()

    
    
def evaluation_metrics(most_suitable_k):
    knn_model=KNeighborsClassifier(n_neighbors=most_suitable_k,weights='uniform',algorithm='auto',p=2)
    knn_model.fit(X_train,Y_train)
    Y_predict=knn_model.predict(X_test)
    print("\nACCURACY SCORE -> ",knn_model.score(X_test,Y_test))

    CM=confusion_matrix(Y_test,Y_predict)
    print("\nCONFUSION MATRIX: \n",CM)
    TNR=CM[0][0]/(CM[0][1]+CM[0][0])
    TPR=CM[1][1]/(CM[1][1]+CM[1][0])

    print("\nEVALUATION:")
    table2=Texttable()
    table2.add_rows([["True_Positive_Rate","True_Negative_Rate","Precision_Score","F-Score "],
                 [TPR,TNR,precision_score(Y_test,Y_predict),f1_score(Y_test,Y_predict)]])
    print(table2.draw(),"\n\n")

    
plot_train_test_error(train_error1,test_error1)

evaluation_metrics(most_suitable_k)
```

**III.** Since the computation time depends on the size of the training set, one may
only use a subset of the training set. Plot the best test error rate,2 which
is obtained by some value of k, against the size of training set, when the
size of training set is N {10, 20, 30,.., 210}. Note: for each N, select
your training set by choosing the first N/3 rows of Class 0 and the first
(N  N/3) rows of Class 1 in the training set you created in 1(b)iii. Also, for
each N, select the optimal k from a set starting from k = 1, increasing by 5.
For example, if N = 200, the optimal k is selected from { 6, 11, . ,196}.
This plot is called a Learning Curve.

```python
def plot_learning_curve():
    print("LEARNING CURVE PLOT: \n")
    optimal_k=1
    optimal_score=0
    score=0
    train_subset_size=np.arange(10,211,10)
    optimal_test_error_rates=empty([21])
    varying_train_size=[i for i in range(10,211,10)]
    for i,N in enumerate(train_subset_size):
        optimal_k=1
        optimal_score=0
        score=0
        train_subset=pd.concat([DataTrain.query('Class==0').head(int(N/3)),DataTrain.query('Class==1').head(int(N-N/3))],
                            ignore_index=True)
        x_train=train_subset[['pelvic incidence','pelvic tilt',
                          'lumbar lordosis angle','sacral slope','pelvic radius','grade of spondylolisthesis']]
        y_train=train_subset['Class']
        y_train=y_train.astype('int')
        for k in range(1,N,5):
            knn_model=KNeighborsClassifier(n_neighbors=k,algorithm='auto',weights='uniform',p=2)
            knn_model.fit(x_train,y_train)
            Y_predict=knn_model.predict(X_test)
            score=knn_model.score(X_test,Y_test)
            if(score>optimal_score):
                optimal_score=score
                optimal_k=k       
        optimal_test_error_rates[i]=(1-optimal_score)*100

    plt.title("LEARNING CURVE")
    plt.plot(varying_train_size,optimal_test_error_rates,color='orange',label="BEST TEST ERROR RATES")
    plt.legend()
    plt.xlabel("varying training size(N)")
    plt.ylabel("error rates")
    plt.show()


    
plot_learning_curve()
print("\n\n")
```

**(d)** Replace the Euclidean metric with the following four metrics and test them. Summarize the
test errors (i.e., when k = k) in a table. Use all of your training data and select the best k when
{ 1, 6, 11,.,196}.
1.  Minkowski Distance:
    1.  which becomes Manhattan Distance with p = 1.
    1.  with log10(p) ={ 0.1, 0.2, 0.3,., 1}. In this case, use the k you found for the Manhattan
    distance in 1(d)i(A). What is the best log10(p)?
    1.  which becomes Chebyshev Distance with p->inifinity
1.  Mahalanobis Distance

```python
import math

neighbors_size=np.arange(1,197,5)

train_error2=np.empty([int(196/5)+1])
train_error3=np.empty([10])
train_error4=np.empty([int(196/5)+1])
train_error5=np.empty([int(196/5)+1])


print("-------KNN using Different Distance Metrics----------\n")


def KNN_manhattan_dist():
    print("KNN using Manhattan distance: ")
    maxScore1=0
    score1=0
    most_suitable_k1=1
    for i,k in enumerate(neighbors_size):
        knn_model=KNeighborsClassifier(n_neighbors=k,algorithm='auto',weights='uniform',p=1)
        knn_model.fit(X_train,Y_train)
        score1=knn_model.score(X_train,Y_train)
        train_error2[i]=(1-score1)*100
        score1=knn_model.score(X_test,Y_test)
        Y_predict=knn_model.predict(X_test)
        if(score1>maxScore1):
            maxScore1=score1
            most_suitable_k1=k       
    table4=Texttable()
    table4.add_rows([["Max Score","Most Suitable K"],[maxScore1,most_suitable_k1]])
    print(table4.draw(),"\n")
    return maxScore1,most_suitable_k1,train_error2




def KNN_minkowski(most_suitable_k1):
    print("KNN using Minkowski distance (using optimal k value from KNN using Manhattan distance) : ")
    p_size=np.arange(0.1,1,0.1)
    maxScore2=0
    score2=0
    new_p=0.1
    best_p=0
    index=0
    for index,new_p in enumerate(p_size):
        knn_model=KNeighborsClassifier(n_neighbors=most_suitable_k1,algorithm='auto',weights='uniform',p=pow(10,new_p))
        knn_model.fit(X_train,Y_train)
        Y_predict=knn_model.predict(X_test)
        score2=knn_model.score(X_train,Y_train)
        train_error3[index]=(1-score2)*100
        score2=knn_model.score(X_test,Y_test)
        if(score2>maxScore2):
            maxScore2=score2
            best_p=new_p
        new_p=new_p+0.1      
    table5=Texttable()
    table5.add_rows([["Max Score","Most Suitable K","Best p"],[maxScore2,most_suitable_k1,best_p]])
    print(table5.draw(),"\n")
    return maxScore2,train_error3




def KNN_chebyshev_dist():
    print("KNN using Chebyshev Distance : ")
    maxScore3=0
    score3=0
    most_suitable_k3=0
    for i,k in enumerate(neighbors_size):
        knn_model=KNeighborsClassifier(n_neighbors=k,metric='chebyshev',p=math.inf)
        knn_model.fit(X_train,Y_train)
        Y_predict=knn_model.predict(X_test)
        score3=knn_model.score(X_train,Y_train)
        train_error4[i]=(1-score3)*100
        score3=knn_model.score(X_test,Y_test)
        if(score3>maxScore3):
            maxScore3=score3
            most_suitable_k3=k      
    table6=Texttable()
    table6.add_rows([["Max Score","Most Suitable K"],[maxScore3,most_suitable_k3]])
    print(table6.draw(),"\n")
    return maxScore3,most_suitable_k3,train_error4




def KNN_mahalanobis_dist():
    print("KNN using Mahalanobis distance: ")
    maxScore4=0
    most_suitable_k4=1
    score4=0
    inv_cov=np.linalg.inv(X_train.cov()) ##inverse covariance of X_train
    for i,k in enumerate(neighbors_size):
        knn_model=KNeighborsClassifier(n_neighbors=k,weights='uniform',algorithm='auto',
                                   metric='mahalanobis',metric_params={'V':inv_cov})
        knn_model.fit(X_train,Y_train)
        Y_predict=knn_model.predict(X_test)
        score4=knn_model.score(X_train,Y_train)
        train_error5[i]=(1-score4)*100
        score4=knn_model.score(X_test,Y_test)
        if(score4>maxScore4):
            maxScore4=score4
            most_suitable_k4=k

    table6=Texttable()
    table6.add_rows([["Max Score","Most Suitable K"],[maxScore4,most_suitable_k4]])
    print(table6.draw(),"\n")
    return maxScore4,most_suitable_k4,train_error5




def summarize_knn_distance_metrics():
    maxScore1,most_suitable_k1,train_error2=KNN_manhattan_dist()
    maxScore2,train_error3=KNN_minkowski(most_suitable_k1)
    maxScore3,most_suitable_k3,train_error4=KNN_chebyshev_dist()
    maxScore4,most_suitable_k4,train_error5=KNN_mahalanobis_dist()
    
    testError1=(1-maxScore1)*100
    testError2=(1-maxScore2)*100
    testError3=(1-maxScore3)*100
    testError4=(1-maxScore4)*100

    print("Summary of KNN with four distance metrics: ")
    table7=Texttable()
    table7.add_rows([["DISTANCE","Best K","Best Test Error Rate"],
                 ['MANHATTAN',most_suitable_k1,testError1],
                 ['MINKOWSKI(log10p)',most_suitable_k1,testError2],
                 ['CHEBYSHEV',most_suitable_k3,testError3],
                 ['MAHALANOBIS',most_suitable_k4,testError4]])

    print(table7.draw(),"\n")
    return train_error2,train_error3,train_error4,train_error5
    
    


train_error2,train_error3,train_error4,train_error5=summarize_knn_distance_metrics()
print("\n\n")
```

**(e)** The majority polling decision can be replaced by weighted decision, in which the
    weight of each point in voting is inversely proportional to its distance from the
    query/test data point. In this case, closer neighbors of a query point will have
    a greater influence than neighbors which are further away. Use weighted voting
    with Euclidean, Manhattan, and Chebyshev distances and report the best test
    errors when k { 6, 11, 16,., 196}.

```python
neighbors_size=np.arange(1,197,5)

train_error6=np.empty([int(196/5)+1])
train_error7=np.empty([int(196/5)+1])
train_error8=np.empty([int(196/5)+1])

print("-------KNN Using MAJORITY POLLING---------\n")

def KNN_majpoll_eucledian():
    print("EUCLEDIAN DISTANCE: ")
    maxScore1=0
    most_suitable_k1=1
    score1=0
    for i,k in enumerate(neighbors_size):
        knn_model=KNeighborsClassifier(n_neighbors=k,algorithm='auto',p=2,weights='distance')
        knn_model.fit(X_train,Y_train)
        Y_predict=knn_model.predict(X_test)
        score1=knn_model.score(X_train,Y_train)
        train_error6[i]=(1-score1)*100
        score1=knn_model.score(X_test,Y_test)
        if(score1>maxScore1):
            maxScore1=score1
            most_suitable_k1=k
    table1=Texttable()
    table1.add_rows([['Max Score','Most Suitable K'],[maxScore1,most_suitable_k1]])
    print(table1.draw(),"\n")
    return maxScore1,most_suitable_k1,train_error6




def KNN_majpoll_manhattan():
    print("MANHATTAN DISTANCE: ")
    maxScore2=0
    most_suitable_k2=1
    score2=0
    for i,k in enumerate(neighbors_size):
        knn_model=KNeighborsClassifier(n_neighbors=k,p=1,algorithm='auto',weights='distance')
        knn_model.fit(X_train,Y_train)
        Y_predict=knn_model.predict(X_test)
        score2=knn_model.score(X_train,Y_train)
        train_error7[i]=(1-score2)*100
        score2=knn_model.score(X_test,Y_test)
        if(score2>maxScore2):
            maxScore2=score2
            most_suitable_k2=k
    table2=Texttable()
    table2.add_rows([["Max Score","Most Suitbale K"],[maxScore2,most_suitable_k2]])
    print(table2.draw(),"\n")
    return maxScore2,most_suitable_k2,train_error7



def KNN_majpoll_chebychev():
    print("CHEBYCHEV DISTANCE: ")
    maxScore3=0
    most_suitable_k3=0
    score3=0
    for i,k in enumerate(neighbors_size):
        knn_model=KNeighborsClassifier(n_neighbors=k,p=math.inf,algorithm='auto',weights='distance')
        knn_model.fit(X_train,Y_train)
        Y_predict=knn_model.predict(X_test)
        score3=knn_model.score(X_train,Y_train)
        train_error8[i]=(1-score3)*100
        score3=knn_model.score(X_test,Y_test)
        if(score3>maxScore3):
            maxScore3=score3
            most_suitable_k3=k
    table3=Texttable()
    table3.add_rows([["Max Score","Most Suitable K"],[maxScore3,most_suitable_k3]])
    print(table3.draw(),"\n")
    return maxScore3,most_suitable_k3,train_error8


    
def KNN_summarize_majority_polling():
    maxScore1,most_suitable_k1,train_error6 = KNN_majpoll_eucledian()
    maxScore2,most_suitable_k2,train_error7 = KNN_majpoll_manhattan()
    maxScore3,most_suitable_k3,train_error8 = KNN_majpoll_chebychev()
    
    print("Summary of best test errors and best K with Majority polling\n")
    testError1=(1-maxScore1)*100
    testError2=(1-maxScore2)*100
    testError3=(1-maxScore3)*100

    table4=Texttable()
    table4.add_rows([["DISTANCE","Best Test Error ","Best K"],
                     ['EUCLEDIAN',testError1,most_suitable_k1],
                     ['MANHATTAN',testError2,most_suitable_k2],
                     ['CHEBYCHEV',testError3,most_suitable_k3]])
    print(table4.draw(),"\n")
    
    return train_error6,train_error7,train_error8
    

    
train_error6,train_error7,train_error8=KNN_summarize_majority_polling()
print("\n\n")
```
