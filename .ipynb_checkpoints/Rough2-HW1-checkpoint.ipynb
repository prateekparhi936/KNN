```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
sn.set(style='ticks')
Data = pd.read_csv('HW1_Dataset/vertebral_column_data/column_2C.dat', sep='\s+', header=None, skiprows=0)
Data
Data.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope', 'pelvic radius', 'grade of spondylolisthesis', 'Class']
Data['Class']=np.where(Data["Class"]=="AB",1,Data['Class'])
Data['Class']=np.where(Data['Class']=="NO",0,Data['Class'])
Data = pd.read_csv('HW1_Dataset/vertebral_column_data/column_2C.csv')
```

```python
column_names = ['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle',
                'sacral slope', 'pelvic radius', 'grade of spondylolisthesis']
sn.pairplot(Data,hue='Class',vars=column_names,kind="scatter",palette="husl")
```

```python
#Question 1(b)(ii)
plt.figure(figsize=(24, 24))
for i,col in enumerate(Data.columns):
    if col!='Class':
        plt.subplot(4, 2, i+1)
        sn.boxplot(x = 'Class', y = col , data = Data, hue = 'Class')
```

```python
rowClass0=70
rowClass1=140
nocol=7

trainClass0=np.zeros(shape=(rowClass0,nocol))
a=[]
for i,row in Data.iterrows():
    if(row['Class']==0):
        a.append(row)
        
for i,row in enumerate(a):
    if i==rowClass0:
        break
    trainClass0[i]=row

testClass0=np.zeros(shape=(len(a)-trainClass0.shape[0],nocol))
x=trainClass0.shape[0]
for i,row in enumerate(a):
    if i==len(a):
        break
    if i>=x:
        testClass0[x-i]=row

trainClass0=pd.DataFrame(trainClass0)
trainClass0.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope', 'pelvic radius', 'grade of spondylolisthesis', 'Class']
trainClass0.Class=trainClass0.Class.astype(int)
        
testClass0=pd.DataFrame(testClass0)
testClass0.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope', 'pelvic radius', 'grade of spondylolisthesis', 'Class']
testClass0.Class=testClass0.Class.astype(int)



trainClass1=np.zeros(shape=(rowClass1,nocol))
b=[]
for i,row in Data.iterrows():
    if(row['Class']==1):
        b.append(row)
        
        
for i,row in enumerate(b):
    if i==rowClass1:
        break
    trainClass1[i]=row
    
testClass1=np.zeros(shape=(len(b)-trainClass1.shape[0],nocol))
x=trainClass1.shape[0]
for i,row in enumerate(b):
    if i==len(b):
        break
    if i>=x:
        testClass1[x-i]=row
        
testClass1=pd.DataFrame(testClass1)
testClass1.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope', 'pelvic radius', 'grade of spondylolisthesis', 'Class']
testClass1.Class=testClass1.Class.astype(int)
        
    
trainClass1=pd.DataFrame(trainClass1)
trainClass1.columns=['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope', 'pelvic radius', 'grade of spondylolisthesis', 'Class']
trainClass1.Class=trainClass1.Class.astype(int)


DataTrain=pd.concat([trainClass0,trainClass1],ignore_index=True)
DataTest=pd.concat([testClass0,testClass1],ignore_index=True)

X_train=DataTrain[['pelvic incidence','pelvic tilt','lumbar lordosis angle',
                   'sacral slope','pelvic radius','grade of spondylolisthesis']]
Y_train=DataTrain['Class']
Y_train=Y_train.astype('int')

X_test=DataTest[['pelvic incidence','pelvic tilt',
                 'lumbar lordosis angle','sacral slope','pelvic radius','grade of spondylolisthesis']]
Y_test=DataTest['Class']
Y_test=Y_test.astype('int')
```

```python
# Y_test is our true or actual value(labels) assigned which we will compare or evaluate with our predicted value i.e Y_predict
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from numpy import empty

neighbors=np.arange(210,0,-2)
maxScore = 0
most_suitable_k = 0;
score=0

# Because of step size 3 there are 70 data points
train_error=empty([105])
test_error=empty([105])    

for i,k in enumerate(neighbors):
    
    ## Initialize the knn model with k neighbors and p=2 (Eucledian Distance).
    Knn = KNeighborsClassifier(n_neighbors=k,p=2)
    
    ## Fit the model on tranining data
    Knn.fit(X_train,Y_train)
    
    ## Prediction on  test data
    Y_predict = Knn.predict(X_test)
    
    ## Check how the model performs on the test data.
    score = Knn.score(X_test,Y_test,sample_weight=None)
    
    ## Train and Test error
    test_error[i]=(1-Knn.score(X_test,Y_test,sample_weight=None))*100
    train_error[i]=(1-Knn.score(X_train,Y_train,sample_weight=None))*100
    
    ## find the k value for which we get maximum score
    if score > maxScore:
        maxScore = score
        most_suitable_k = k
        
print("Most suitable K Value: ",most_suitable_k,"\nBest Score: ",score)
plt.title("Plotted Train and Test Errors in terms of Neighbors(K)")
plt.plot(neighbors, test_error, label='Test error')
plt.plot(neighbors, train_error, label='Train error')
plt.legend()
plt.xlabel('NEIGHBORS(K)')
plt.ylabel('ERROR')
plt.show()
```

```python
# Learning Curve
optimal_K=1
optimal_score=0
best_test_error_rate=empty([21])
train_size=np.arange(10,211,10)
varying_N=[i for i in range(10,211,10)]
for i,N in enumerate(train_size):
    train_subset=pd.concat([DataTrain.query('Class==0').head(int(N/3)),DataTrain.query('Class==1').head(int(N-N/3))],ignore_index=True)
    train_x=train_subset[['pelvic incidence', 'pelvic tilt', 'lumbar lordosis angle','sacral slope', 'pelvic radius', 'grade of spondylolisthesis']]
    train_y=train_subset['Class']
    for k in range(1,N,5):
        knn=KNeighborsClassifier(n_neighbors=k,p=2)
        knn.fit(train_x,train_y)
        score=knn.score(X_test,Y_test,sample_weight=None)
        if score>optimal_score:
            optimal_score=score
            optimal_K=k
    best_test_error_rate[i]=(1-optimal_score)

plt.title("LEARNING CURVE")
plt.plot(varying_N,best_test_error_rate,label="BEST TEST ERROR RATE")
plt.legend()
plt.xlabel("varying N(size of training set)")
plt.ylabel("best error rate")
plt.show()     
```
